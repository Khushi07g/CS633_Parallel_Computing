```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Define NP values to consider
np_values = [8, 16, 32, 64]

# --- Data for Test Case 1 ---
data_case1 = {
    8: [
         (0.024100, 0.007836, 0.134730),
         (0.061245, 0.007812, 0.174919),
         (0.147669, 0.009941, 0.269514),
         (0.035090, 0.007830, 0.147230)
       ],
    16: [
         (0.596905, 0.079800, 0.786549),
         (0.558874, 0.106966, 0.763940),
         (0.604408, 0.126968, 0.807231),
         (0.730337, 0.109837, 0.932392)
       ],
    32: [
         (1.419834, 0.318964, 1.719661),
         (1.258020, 0.219985, 1.511722),
         (1.318233, 0.263878, 1.611565),
         (1.384300, 0.267969, 1.723840)
       ],
    64: [
         (0.349063, 0.020591, 0.459671),
         (0.320610, 0.014465, 0.431441),
         (0.289116, 0.066150, 0.399317),
         (0.236651, 0.014225, 0.314858)
       ]
}

# --- Data for Test Case 2 ---
data_case2 = {
    8: [
         (0.033054, 0.025135, 0.058341),
         (0.033278, 0.025060, 0.165621),
         (0.081591, 0.031664, 0.113173),
         (0.139106, 0.031773, 0.269398)
       ],
    16: [
         (0.520485, 0.107887, 0.606572),
         (0.635947, 0.131968, 0.875888),
         (0.597262, 0.105872, 0.712268),
         (0.589193, 0.106944, 0.818246)
       ],
    32: [
         (1.385812, 0.268976, 1.588262),
         (1.382044, 0.255678, 1.652380),
         (1.433774, 0.263964, 1.588196),
         (1.414284, 0.220974, 1.705273)
       ],
    64: [
         (0.298941, 0.042853, 0.319234),
         (0.402264, 0.043158, 0.526404),
         (0.304953, 0.050752, 0.333807),
         (0.377531, 0.047689, 0.513372)
       ]
}

def create_df(data, test_case_label):
    rows = []
    for np_val in np_values:
        runs = data.get(np_val, [])
        for run_id, (read, main, total) in enumerate(runs, start=1):
            rows.append({"TestCase": test_case_label, "NP": np_val, "Run": run_id, "Phase": "Read", "Time": read})
            rows.append({"TestCase": test_case_label, "NP": np_val, "Run": run_id, "Phase": "Main", "Time": main})
            rows.append({"TestCase": test_case_label, "NP": np_val, "Run": run_id, "Phase": "Total", "Time": total})
    return pd.DataFrame(rows)

df_case1 = create_df(data_case1, "Test Case 1")
df_case2 = create_df(data_case2, "Test Case 2")
df_all = [df_case1, df_case2]

def calculate_mean(data):
    mean_data = []
    for np_val, runs in data.items():
        # Calculate mean for each position in the tuple
        mean_read = sum(run[0] for run in runs) / len(runs)
        mean_main = sum(run[1] for run in runs) / len(runs)
        mean_total = sum(run[2] for run in runs) / len(runs)

        mean_data.append({
            "NP": np_val,
            "Mean Read": mean_read,
            "Mean Main": mean_main,
            "Mean Total": mean_total
        })
    return pd.DataFrame(mean_data)

# Plot each phase
phases = ["Read", "Main", "Total"]

for df in df_all:
  i = 1
  plt.figure(figsize=(15,5))
  for phase in phases:
      plt.subplot(1,3,i)
      # Box plot
      sns.boxplot(data=df[df["Phase"] == phase], x="NP", y="Time", hue="TestCase", showcaps=True, width=0.4, fliersize=5, palette="pastel")
      means = df[df["Phase"] == phase].groupby(["NP", "TestCase"])["Time"].mean().reset_index()

      # Add line plots on top of box plots
      for tc in means["TestCase"].unique():
          tc_data = means[means["TestCase"] == tc]
          plt.plot(np.arange(len(tc_data["NP"].unique())), tc_data["Time"],
                  'o-', label=f"Mean {tc}", linewidth=2)

      plt.title(f"{phase} Time vs Number of Processes")
      plt.xlabel("Number of Processes (NP)")
      plt.ylabel("Time (seconds)")
      plt.grid(True, linestyle="--", alpha=0.5)
      plt.legend()
      i += 1

  plt.show()
```
